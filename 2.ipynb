```
# getting income by zip added to housing data
zip_income['zip'] = zip_income['zip'].astype('int')

zip_dict = dict(zip(zip_income.zip, zip_income['AVG(agi)']))
housing_df['avg_income'] = housing_df['zipcode'].map(zip_dict)

# Convert the avg_income from a string to a float
housing_df['avg_income'] = housing_df['avg_income'].astype('float')
```

```
# creating a month column derived from date
housing_df['date'] = pd.to_datetime(housing_df['date'])
housing_df['month'] = housing_df['date'].dt.month
# Add year_sold derived from date
housing_df['year_sold'] = housing_df['date'].dt.year
```

```
# Dropping rows where we are almost positive that there is an error in the number of bedrooms and bathrooms
housing_df = housing_df.drop([6984, 1419, 1996, 3811, 4796, 7673, 8461, 14120])
```

```
# Make zipcode a category and bin it!
# One-hot-encode it where we split it
housing_df['zipcode']= housing_df['zipcode'].astype('category')
#housing_df.info()
#housing_df
```

```
# Scale sqft features
# from sklearn.preprocessing import MinMaxScaler

# scaler = MinMaxScaler() 

# housing_df['sqft_living_scaled'] = scaler.fit_transform(housing_df[ ['sqft_living'] ])
# housing_df['sqft_lot_scaled'] = scaler.fit_transform(housing_df[ ['sqft_lot'] ])
# housing_df['sqft_above_scaled'] = scaler.fit_transform(housing_df[ ['sqft_above'] ])
# Drop sqft_living, sqft_lot and sqft_above when creating dataframe X
```

```
# Split data in X and y
#  'sqft_living', 'sqft_lot', 'sqft_above' if you scale it like above, you need to drop these columns
X = housing_df.drop(['price', 'date', 'id'], axis=1) ### Dropped date but will fix it later in feature engineering in order to include it. It needs to be a number and not an object

# One-hot-encode any categorical (just zipcode)
features = list(X.columns)
X = pd.get_dummies(X[features], drop_first=True)

y = housing_df['price']
```

```
# Split data in Xtest, ytest, Xtrain, ytrain
seed = 7 # Random state seed
test_size = 0.20 # An 80-20 split for train-to-test ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

# Model Engineering

```
# Functions for finding best parameters
def display_scores(scores):
    print("Scores: {0}\nMean: {1:.3f}\nStd: {2:.3f}".format(scores, np.mean(scores), np.std(scores)))

def report_best_scores(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```
model = XGBRegressor()
search = RandomizedSearchCV(model, param_distributions=params, random_state=42, n_iter=200, cv=3, verbose=2, n_jobs=1, return_train_score=True)
search.fit(X, y)
report_best_scores(search.cv_results_, 1)
```

```
# Use the new parameters to create model
new_model = XGBRegressor(objective = 'reg:squarederror', colsample_bytree = 0.776689016483971, learning_rate=0.13069801943908896, max_depth = 6, n_estimators= 145, subsample= 0.9082402575984808)
new_model.fit(X_train, y_train)
```

# Model Evaluation

```
# Get our predictions from the model
predictions = new_model.predict(X_test)
```

```
# Display the metrics for regression

# Get the RMSE (Root Mean Squared Error) and MSE (Mean Squared Error).
# Same function but if squared is False it gives the RMSE and true gives us the MSE
rmse = mean_squared_error(y_test, predictions, squared=False)
mse = mean_squared_error(y_test, predictions, squared=True)

# Get the MAE (Mean absolute Error) and R2-score
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

# Display all
print(f'Root Mean Square Error: {rmse}')
print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R2-score {r2}')
```
